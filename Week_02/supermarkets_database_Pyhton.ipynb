# Writing data to and reading data from a Database using Python (Task 4.f)

## Libraries and settings

```python
# Libraries
import os
import sqlite3
import fnmatch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Function to close a sqlite db-connection
def check_conn(conn):
     try:
        conn.cursor()
        return True
     except Exception as ex:
        return False

# Get current working directory
print(os.getcwd())
```

## Create sqlite data base

```python
# Create data base
conn = sqlite3.connect('supermarkets_database.db') 
cursor = conn.cursor()

# Show dbs in the directory
flist = fnmatch.filter(os.listdir('.'), '*.db')
for i in flist:
    print(i)
```

## Create SQL-table in the database

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS supermarkets_table (Type VARCHAR(50),
                                                               ID INT(10),
                                                               Lat DECIMAL(5,6),
                                                               Lon DECIMAL(5,6),
                                                               Brand VARCHAR(50),
                                                               Shop VARCHAR(50),
                                                               City VARCHAR(50),
                                                               Street VARCHAR(50),
                                                               Housenumber VARCHAR(10),
                                                               Postcode VARCHAR(10))''')
# Confirm changes to the table
conn.commit()
```

## Read data from file to data frame

```python
df = pd.read_csv('supermarkets_data_prepared.csv', 
                  sep=',', 
                  encoding='utf-8')[['type','id','lat','lon','brand','shop','city','street','housenumber','postcode']]
print(df.shape)
df.head(5)
```

## Write data to the SQL-table in data base

```python
df.to_sql(name = 'supermarkets_table',
          con = conn,
          index = False,
          if_exists = 'replace')
```

## Query the SQL-table

```python
# Query the SQL-table
cursor.execute('''SELECT *
               FROM supermarkets_table
               WHERE City='Winterthur' ''')

df = pd.DataFrame(cursor.fetchall(), 
                  columns=['Type','Id','Lat','Lon','brand','Shop','City','Street','Housenumber','Postcode'])    
df
```

## Close db connection (if open)

```python
# Close db connection (if open)
try:
    if check_conn(conn):
        conn.close()
    else:
        pass
except:
    pass

# Status (True = open, False = closed)
print(check_conn(conn))
```
